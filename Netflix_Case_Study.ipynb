# **Netflix Analysis and Visualisation**

## **Problem Definition** - Analyze the data and generate insights that could help Netflix in deciding which type of shows/movies to produce and how they can grow the business in different countries**

The dataset provided to you consists of a list of all the TV shows/movies available on Netflix:

**Show_id**: Unique ID for every Movie / Tv Show

**Type**: Identifier - A Movie or TV Show

**Title**: Title of the Movie / Tv Show

**Director**: Director of the Movie

**Cast**: Actors involved in the movie/show

**Country**: Country where the movie/show was produced

**Date_added**: Date it was added on Netflix

**Release_year**: Actual Release year of the movie/show

**Rating**: TV Rating of the movie/show

**Duration**: Total Duration - in minutes or number of seasons

**Listed_in**: Genre

**Data Cleaning**

```python
# checking null values in columns
df.isnull().sum()
```

```python
# Checking percentage of null values in column
null_value_count = df.isnull().sum()
total = len(df)
print('Percentage of null values in columns :')
for column in df:
  print(f'{round((null_value_count[column]/total)*100, 2)}% - {column}')
```

conclusion
*   Director column has approx. 30% of null values.
*   duration, rating, dated_added column has less than 1% of null values, so we can drop those.



```python
# filling missing values
df['director']=df['director'].fillna('NA') # director column
df['cast']=df['cast'].fillna('NA') # cast column
df['country']=df['country'].fillna('NA') # country column
```

```python
# shifting duration values from rating column to duration column
df.loc[df['rating'].str.contains('min')==True, 'duration'] = df.loc[df['rating'].str.contains('min')==True, 'rating']

# replacing missing values of rating column to mode values
df.loc[df['rating'].str.contains('min')==True, 'rating'] = df['rating'].mode()[0]
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])

# result after replacing
print(df['rating'].unique())
```

```python
# getting values by discarding null values of 'date_added' column
df = df.loc[df['date_added'].isnull()==False]

# final data without any null values
df.isnull().sum()
```

```python
# changing data types
df['date_added'] = pd.to_datetime(df['date_added'])
df = df.astype({'type':'category', 'rating':'category'})
```

```python
# final shape of data
df.shape
```

# **Univariate Analysis**

```python
plt.figure(figsize = (10,5))
sns.countplot(df['type'])
plt.title('Count of Movies/TV Shows on Netflix')
plt.show()
```

conclusion
*   There are more number of movies on Netflix as compared to tv shows.
*   Therefore, Netflix has more focus on Movies.



```python
plt.figure(figsize=(20,5))

df_year = df.loc[df['release_year']>2000]

# getting year-wise count of movies and tv showa
df_year_wise_count = df.groupby('release_year')['type'].value_counts().reset_index()
df_year_wise_count = df_year_wise_count.loc[df_year_wise_count['release_year']>2000]
df_year_wise_count.rename(columns = {'level_1':'type', 'type':'count'}, inplace=True)

# subplots
#bar plot
plt.subplot(1,2,1)
sns.countplot(x='release_year', data = df_year, hue='type')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.title('Year wise count of Movies/TV Shows')

#line plot
plt.subplot(1,2,2)
sns.lineplot(data=df_year_wise_count, x = 'release_year', y='count', hue='type' )
plt.xticks(df_year_wise_count['release_year'], rotation=45)
plt.title('Year wise Movie/TV Show count')

#show plot
plt.show()
```

conclusion
*   More number of movies are added till 2018 as compared to tv shows but number of tv shows has exceeded movies in recent year.
*   There is sudden decrease in movies from 2018 to 2021 but no. of tv shows kept on increasing.  



```python
# Extracting days, months and year from date_added column
time_df = df.copy()
time_df['year'] = time_df['date_added'].dt.year
time_df['month'] = time_df['date_added'].dt.month
time_df['day'] = time_df['date_added'].dt.day_name()
month_df = time_df.sort_values(by='month')
month_df['month_name'] = month_df['date_added'].dt.month_name()

# setting figure size
plt.figure(figsize=(12,5))
# count plot
sns.countplot(x='month_name', data=month_df, hue='type')
plt.xticks(rotation=45) # rotating x values to 45 degree
plt.legend(loc=(1.01,0.5)) # legend
plt.title('Movies/TV Show added per month on Netflix') # setting plot title
plt.show() # show
```

conclusion
*   More content is added on January, April, July, October, December. Preferably holiday months. 




```python
# setting figure size
plt.figure(figsize=(10,5))
# count plot
sns.countplot(x='day', data=time_df, hue='type')
plt.title('Movie/TV Show added per day') # title
plt.show() # show
```

conclusion
*   More content is added on Friday (as the weekend approaches).



```python
# Extracting season data
df_seasons = df.loc[df['duration'].str.contains('Season')==True, 'duration'].value_counts().reset_index()[:10]
df_seasons.rename(columns = {'index':'season', 'duration':'count'}, inplace=True)

# Extracting minutes data
duration_df = df.loc[df['duration'].str.contains('min')==True, 'duration'].apply(lambda x: x.split()[0]).astype(int)

# setting figure size
plt.figure(figsize = (18,5))
# creating subplots
# plotting season
plt.subplot(1,2,1)
sns.barplot(x='count', y='season', data=df_seasons, color='firebrick')
plt.title('Seasons on Netflix')

#plotting duration
plt.subplot(1,2,2)
sns.boxplot(duration_df, color='firebrick')
plt.title('Maximum number of duration on Netflix ')

plt.show()# show
```

conclusion
*   Mostly season 1 shows are available.
*   Maximum close to 100 mins shows are available on Netflix.

## **Bi-Variate Analysis**

```python
# Extracting movie-tv show count
movie_content = df.loc[df['type']=='Movie', ['type', 'rating']]
tv_show_content = df.loc[df['type']=='TV Show', ['type', 'rating']]

# setting figure size
plt.figure(figsize = (18,5))

#plot subplot
plt.subplot(1,2,1)
sns.countplot(y='rating', data=movie_content, color='royalblue')
plt.title('Movie-Content Count')

plt.subplot(1,2,2)
sns.countplot(y='rating', data=tv_show_content, color="royalblue")
plt.title('TV Show-Content Count')


plt.show()
```

conclusion
*   Most of the content is available for adults both in case of movies and TV shows.
*   Decent amount of content is available for teenagers.
*   Less amount of content is available for kids.






```python
# unnesting country column

country_df = df["country"].apply(lambda x: str(x).split(", ")).tolist()
country_df = pd.DataFrame(country_df, index = df["title"])
country_df = country_df.stack()
country_df = country_df.reset_index()
country_df.drop(columns = "level_1" , inplace = True)
country_df.columns = ["title" , "country"]
country_df = df.merge(country_df , on = "title")
country_df.drop(columns ="country_x" , inplace = True)
country_df.rename(columns = {"country_y" :"country"}, inplace = True)
country_df.head(3)
```

```python
#Extracting movie and tv show count country wise

#Movie
country_movie_count = country_df.loc[country_df['type']=='Movie', 'country']
country_movie_count = country_movie_count.value_counts().reset_index()[:20]
country_movie_count.rename(columns = {'index':'Country', 'country':'count'}, inplace=True)
country_movie_count = country_movie_count.loc[country_movie_count['Country']!='NA']

#TV Show
country_tvshow_count = country_df.loc[country_df['type']=='TV Show', 'country']
country_tvshow_count = country_tvshow_count.value_counts().reset_index()[:20]
country_tvshow_count.rename(columns = {'index':'Country', 'country':'count'}, inplace=True)
country_tvshow_count = country_tvshow_count.loc[country_tvshow_count['Country']!='NA']

#Plotting
plt.figure(figsize=(10,10))

# subplots
plt.subplot(2, 1, 1)
sns.barplot(y='Country', x='count', data = country_movie_count, color='firebrick')
plt.title('Movie Count Country-Wise') # title

plt.subplot(2,1,2)
sns.barplot(y='Country', x='count', data = country_tvshow_count, color='firebrick')
plt.title('TV Show Count Country-Wise') # title

plt.show() # show plot
```

conclusion
*   Ananlysis of top 20 countries is done based on movies and tv shows.
*   United States has highest number of movies and tv shows on Netflix.
*   United Kingdom has slightly more number of tv shows than movies.
*   India has more movies but less tv shows.
*   Japan and South Korea mostly focuses on tv shows as compared to movie.








```python
# unnesting genre 
genre_df = df["listed_in"].apply(lambda x: str(x).split(", ")).tolist()
genre_df = pd.DataFrame(genre_df, index = df["title"])
genre_df = genre_df.stack()
genre_df = genre_df.reset_index()
genre_df.drop(columns = "level_1" , inplace = True)
genre_df.columns = ["title" , "genre"]
genre_df = country_df.merge(genre_df , on = "title")
genre_df.drop(columns ="listed_in" , inplace = True)
genre_df.head(3)
```

```python
#Extracting top 20 genres on Netflix
top_20_genre = genre_df['genre'].value_counts()[:20].reset_index()
top_20_genre.rename(columns = {'index':'genre', 'genre':'count'}, inplace=True)

#setting figure size
plt.figure(figsize=(10,8))
sns.barplot(y='genre', x = 'count', data=top_20_genre, color='firebrick')# bar plot
plt.title('Top 20 Genre on Netflix') # title
plt.show() #show
```

```python
# Extracting genre based on Movies and TV Show

#Movie
genre_movie_df = genre_df.loc[genre_df['type']=='Movie', 'genre'].value_counts()[:10].reset_index()
genre_movie_df.rename(columns = {'index':'Genre', 'genre':'Count'}, inplace=True)

#TV Show
genre_tvshow_df = genre_df.loc[genre_df['type']=='TV Show', 'genre'].value_counts()[:10].reset_index()
genre_tvshow_df.rename(columns = {'index':'Genre', 'genre':'Count'}, inplace=True)

#setting figure size
plt.figure(figsize=(10,10))
# subplots
plt.subplot(2,1,1)
sns.barplot(x='Count', y='Genre', data = genre_movie_df, color='firebrick')
plt.title('Top 10 Movie Genres')

plt.subplot(2,1,2)
sns.barplot(x='Count', y='Genre', data = genre_tvshow_df, color='firebrick')
plt.title('Top 10 TV Show Genres')

plt.show()#show
```

conclusion
*   Most popular genres among movies and tv shows are drama, comedy and international movies and tv shows.




```python
# Extracting top 10 genre for top 10 country
# genre
top_10_genre = genre_df['genre'].value_counts()[:10].reset_index()
top_10_genre.rename(columns = {'index':'genre', 'genre':'count'}, inplace=True)
genre_list = top_10_genre['genre'].tolist()

#country
top_10_country = genre_df.loc[genre_df['country']!='NA', 'country'].value_counts()[:10].reset_index()
top_10_country.rename(columns = {'index':'country', 'country':'count'}, inplace=True)
country_list = top_10_country['country'].tolist()

genre_country_df = genre_df.loc[genre_df['genre'].isin(genre_list)]
genre_country_df = genre_country_df.loc[genre_country_df['country'].isin(country_list)]

genre_country_df = pd.DataFrame(genre_country_df.groupby('country')['genre'].value_counts())
genre_country_df.rename(columns = {'genre':'count'}, inplace=True)
genre_country_df.reset_index(inplace=True)
genre_country_df = genre_country_df.pivot(index='country', columns='genre', values='count')
genre_country_df
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(genre_country_df, annot=True, cmap="Blues", fmt='d')
plt.title('genre analysis country wise')
plt.show()
```

conclusion
*   Comparing top 10 genres with top 10 countries.
*   United State has all genres popular as compared to other countries except for International TV Shows or International Movies.
*   India - Interntional Movies, Dramas and comedies more popular.
*   Japan and South KoreA has more focus on International TV Shows.
*   United Kingdom has more focus on International TV Shows, International Movies, Dramas.
